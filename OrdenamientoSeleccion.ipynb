```python
#ORDENAMIENTO POR SELECCION
import numpy as np;
import time, random;

vt= []
def seleccion(A):
    ti = int(round(time.time()*10000))
    for i in range(len(A)):
        minimo=i;
        for j in range(i,len(A)):
            if(A[j] < A[minimo]):
                minimo=j;
        if(minimo != i):
            aux=A[i];
            A[i]=A[minimo];
            A[minimo]=aux;
    print ("Vector arreglado: \n",A);
    tf = int(round(time.time()*10000))
    t = tf- ti
    vt.append(t);

vn =[50,100,500,1000]
for n in vn:
  A=np.random.randint(1,1000,n);
  print("ARREGLO NORMAL: \n",A);
  seleccion(A);

print ('TIEMPO:\n',vt)
```

```python
import matplotlib.pyplot as plt
plt.plot(vn,vt)
```
