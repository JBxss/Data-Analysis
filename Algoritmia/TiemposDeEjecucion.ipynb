```python
import random
import time

#n = int(input('Cantidad de estudiantes: '))
m = int(input('cantidad de asignaturas: '))

vt = []
vn = [10, 100, 200, 300, 500, 600, 700] #cantidades de estudiantes
ve = [10,100,200,300,500,600,700] # valores de referencia O(n)

for n in vn:
  ti= int(round(time.time()*10000))

  for est in range (1,n+1):
    for asig in range (1,m+1) :
      p1 = (random.uniform(1,5))
      p2 = (random.uniform(1,5))
      p3 = (random.uniform(1,5))
      prom = (p1 + p2 + p3)/3
      #print('Estudiante ', est, ' Asignatura: ', asig,' Promedio: ', prom)

  tf = int(round(time.time()*10000))
  t = tf - ti
  print('Tiempo de ejecución: ', t)
  vt.append(t)
```

```python
from matplotlib import pyplot

pyplot.plot(vn,vt)
pyplot.plot(vn,ve)
```
