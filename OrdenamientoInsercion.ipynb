```python
#ORDENAMIENTO POR INSERCION
import numpy as np;
import time, random;

vt= []
def insercion(A):
    ti = int(round(time.time()*10000))
    for i in range(len(A)):
        for j in range(i,0,-1):
            if(A[j-1] > A[j]):
                aux=A[j];
                A[j]=A[j-1];
                A[j-1]=aux;
    print ('VECTOR ARREGLADO:\n',A);
    tf = int(round(time.time()*10000))
    t = tf- ti
    vt.append(t);

vn =[50,100,500,1000]
for n in vn:
    A=np.random.randint(1,1000,n);
    print ('VECTOR ORIGINAL:\n', A);
    insercion(A);

print ('TIEMPO:\n',vt)
```

```python
import matplotlib.pyplot as plt
plt.plot(vn,vt)
```
